```python
#desenvolva um programa que calcule e exiba a média aritmética de N notas fornecidas pelo usuário.
#Até o usuário desejar interromper o processo
nota = 1
qtdNotas = 0
calculo = 0

#enquanto nota for diferente de -1 (número utilizado para encerrar o programa)
while nota != -1:
  #pede e atribui a nota para a variável
  nota = float(input("Digite a nota (-1 para encerrar):\n"))
  #se a nota for diferente do número utilizado para encerrar, realiza o cálculo
  if nota != -1:
    #contador de quantidade de notas para fazer a divisão no final
    qtdNotas += 1
    #aoma os novos valores digitados e sobrepõe o cálculo seguido
    calculo = nota + calculo
#se a quantidade de notas for diferente de 0
if qtdNotas != 0:
  #imprime na tela o resultado final, a soma das notas divididas pela quantidade
  #de notas
  print("A média aritmética é:", calculo / qtdNotas)
else:
  #se a quantidade de notas for 0, não existe divisão por 0, então imprime
  #"cálculo inválido"
  print("Cálculo inválido.")
```

```python
biblioteca = []

continuar = int(input("Bem-vindo(a) a sua biblioteca! Digite 1 para entrar no sistema ou 2 para sair\n"))
while continuar == 1:
  menu = int(input("Digite a sua escolha:\n1- Adicionar um livro\n2- Remover um livro\n3- Imprimir todos os livros\n"))
  if menu == 1:
    livro = {}
    titulo = input("Digite o título do livro:\n")
    autor = input("Digite o autor do livro:\n")
    ano = int(input("Digite o ano do livro:\n"))
    livro.update(titulo = titulo, autor = autor, ano = ano)
    biblioteca.append(livro)
  elif menu == 2:
    removerLivro = input("Digite o título do livro que deseja remover:\n")
    for livro in biblioteca:
      if livro.get('titulo').lower() == removerLivro.lower():
        biblioteca.remove(livro)
      else:
        print("Não existe nenhum livro com esse título.")
        break
    print(biblioteca)
  elif menu == 3:
    print("Livros disponíveis na sua biblioteca:")
    for liv in biblioteca:
      print(f" Título: {liv['titulo']}, Autor: {liv['autor']}, Ano: {liv['ano']}.")
  else:
    print("Número inválido")
    break
  continuar = int(input("Digite 1 para continuar no sistema ou 2 para sair\n"))
```

```python
#cálculos matemáticos com vetores
import numpy as np

a = np.array([8, 4, 9, 2])
b = np.array([2, 4, 3, 2.5])
print("Vetor a:",a, "\nVetor b:", b)

print("\na + b =")
c = np.add(a, b) #mesma coisa que c = a + b
print(c)

print("\na - b =")
c = np.subtract(a,b) #mesma coisa que c = a - b
print(c)

print("\na * b =")
c = np.multiply(a, b) #mesma coisa que c = a * b
print(c)

print("\na / b =")
c = np.divide(a, b) #mesma coisa que c = a / b
print(c)

print("\na ** b =")
c = np.power(a, b) #mesma coisa que c = a ** b (exponenciação/ a elevado a b)
print(c)

print("\nRaiz quadrada de a =")
c = np.sqrt(a) #mesma coisa que c = raiz quadrada de a
print(c)
```

divisões das variáveis

**categóricas**: representam tipos de dados divididos em grupos ou categorias distintas com número finito de valores, não quantificáveis numericamente. Elas descrevem características como gênero, cor ou tipo sanguíneo. Podem ser **nominais** (sem ordem) ou **ordinais** (com ordem lógica).

**numéricas**: são características mensuráveis representadas por números que descrevem "quanto" ou "quantos" de algo existe. Elas permitem operações aritméticas (soma, média) e dividem-se em **contínuas** (valores decimais/medidas que geralmente são medidas através de algum instrumento, ex: altura, peso) e **discretas** (valores inteiros/contagens, ex: número de filhos, idade em anos).

Medidas de dispersão ou variabilidade

as medidas de dispersão ou de variabilidade têm como objetivo indicar o grau de variação dos elementos de um conjunto numérico em relação a sua média.

**amplitude:** a amplitude é a mais simples das medidas de dispersão e consiste basicamente na diferença entre o maior e o menor valor de um conjunto de dados. Exemplo: amplitude térmica (maior temperatura do dia, subtraído pela menor temperatura do mesmo dia).

**variância:** a variância é a medida que nos diz o quão distante da média estão os nossos dados. Ela é calculada como a média dos quadrados das diferenças entre cada valor e a média.

**desvio padrão:** o desvio padrão é uma medida que expressa o grau de dispersão de um conjunto de dados. Ou seja, o desvio padrão indica o quanto um conjunto de dados é uniforme (quanto mais próximo de 0 for o desvio de dados, mais homogêneo são os dados). Ele é calculado a partir da variância, ao ser a raiz quadrada desse parâmetro. Um desvio padrão baixo indica que os dados estão mais próximos da média, e o contrário indica que estão longe da média.

```python
import pandas as pd
df = pd.read_csv("/content/vinhos - vinhos.csv")
#prática das medidas:
#amplitude (utilizada geralmente no início da análise, para identificar se existem valores extremos)
amp_price = df['value_usd'].max() - df['value_usd'].min() #calcula a amplitude pegando o maior preço e subtraindo pelo menor
amp_rel = amp_price / df['value_usd'].mean() * 100 #calcula a amplitude relativa pegando a amplitude real e dividindo pela
#porcentagem da média (devolve a porcentagem da amplitude em relação a média (nesse caso, a amplitude é 4x maior q a média))
print(f"Amplitude de value_usd: {round(amp_price, 2)}")
print(f"\nAmplitude relativa de value_usd: {round(amp_rel, 2)}")

#variância
#variância da população que escolhemos (value nesse caso)
var_pop = df['value_usd'].var(ddof = 0) #ddof = 0 -> indica que pegamos todos os dados da coluna, se quiser somente uma amostra usar 1
print(f"\nVariância de value_usd: {round(var_pop, 2)}") #nesse caso, a variância é muito alta, ou seja
#uma dispersão de dados muito grande em relação a média dos preços do vinho

#desvio padrão
std_pop = df['value_usd'].std(ddof = 0)
print(f"\nDesvio padrão de value_usd: {round(std_pop, 2)}") #nesse caso, o desvio é muito alto, ou seja,
#uma dispersão de dados muito grande em relação a média dos preços do vinho
```

```python
#exercícios
#1 calcule a amplitude absoluta, relativa, variança e desvio padrão de volume_hl por grape_type. Qual
#a interpretação que teremos?
grupo = df.groupby('grape_type')
amp_abs = grupo['volume_hl'].max() - grupo['volume_hl'].min()
amp_rel = amp_abs / grupo['volume_hl'].mean() * 100
variancia = grupo['volume_hl'].var(ddof = 0)
desvio = grupo['volume_hl'].std(ddof = 0)
print(f"Amplitude Absoluta: {round(amp_abs, 2)}")
print(f"\nAmplitude Relativa: {round(amp_rel, 2)}")
print(f"\nVariância: {round(variancia, 2)}")
print(f"\nDesvio: {round(desvio, 2)}")

#o tipo de uva tem alta dispersão com a quantidade de volume, o desvio padrão é bem alto

#2 selecione uma amostra de 25%, e calcule as medidas de dispersão para value_usd e volume_hl por grape_type.
#Compare-as com as da população. Por que diferem?
amostra = df.sample(frac = 0.25, random_state = 42)
grupo_amostra = amostra.groupby('grape_type')
amp_abs_amostra = grupo_amostra['volume_hl'].max() - grupo_amostra['volume_hl'].min()
media_amostra = grupo_amostra['volume_hl'].mean()
amp_rel_amostra = (amp_abs_amostra / media_amostra) * 100
variancia_amostra = grupo_amostra['volume_hl'].var(ddof = 0)
desvio_amostra = grupo_amostra['volume_hl'].std(ddof = 0)
resultado_amostra = pd.DataFrame({
    "Amplitude Absoluta":amp_abs_amostra,
    "Amplitude Relativa":amp_rel_amostra,
    "Variancia":variancia_amostra,
    "Desvio Padrão":desvio_amostra
})
print("\n",resultado_amostra.round(2))
```

Intervalo Interquartil

Q1 = 0%-25% (menores valores)

Q2 = 25%-50% (média dos valores)

Q3 = 50%-75% (média maior)

Q4 = 75%-100% (maiores valores)

IQR = Q3 - Q1

Quanto maior o IQR (quanto mais distante de 0), maior a dispersão de valores



```python
#visualização de dispersão dos dados
import pandas as pd
df = pd.read_csv("/content/vinhos - vinhos.csv")
q1 = df['value_usd'].quantile(0.25)
q3 = df['value_usd'].quantile(0.75)
iqr = q3-q1
limite_inferior = q1 - 1.5 * iqr #1.5 pega o menor valor aceitavel
limite_superior = q3 + 1.5 * iqr #1.5 pega o menor valor aceitavel

#valore negativo significa variação muito grande
print('Limite inferior:', limite_inferior)
print('\nLimite superior:', limite_superior)

#Z-Score = score padrão, baseado em desvio padrão: + acima da média, - abaixo da média
#z = valor - a media / devio padrao
df["zscore"] = (df['value_usd'] - df['value_usd'].mean()) / df['value_usd'].std(ddof = 0)
print("\n\n",df)
```

```python
#exercícios
#1 calcule o IQR e Z-Score de volume_hl por grape_type e identifique onde se encontra os
#outliers. Qual a interpretação que teremos?
print("Exercício 1:\n")
q1 = df.groupby('grape_type')['volume_hl'].quantile(0.25)
q3 = df.groupby('grape_type')['volume_hl'].quantile(0.75)
iqr = q3 - q1
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 - 1.5 * iqr
print('Limite inferior:', limite_inferior)
print('\nLimite superior:', limite_superior)
outliers_iqr = df[
                  (df['volume_hl'] < df['grape_type'].map(limite_inferior)) |
                  (df['volume_hl'] > df['grape_type'].map(limite_superior))
                 ]
print("\n",outliers_iqr,"\n")
print("*"*60)

#2 selecione uma amostra de 25%, e IQR e Z-Score de volume_hl por grape_type e identifique
#onde se encontra os outliers. Compare-as com as da população. Por que diferem?
```
